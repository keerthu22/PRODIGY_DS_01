```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('world_demographics.csv')
```

```python
print(df.head())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=10, kde=False, color='skyblue')

plt.title('Age Distribution in Sample Population')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
